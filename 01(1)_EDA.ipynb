```python
import pandas as pd

df = pd.read_csv("data.csv")

# Shape
print("Dataset shape:", df.shape)

# Columns & types
print(df.info())

# Basic statistics
print(df.describe())

# First 5 rows
print(df.head())
print(df.isnull())
print(df.count())
```

```python
# Count of fraud vs legit
print(df['is_fraud'].value_counts())
print(df['is_fraud'].value_counts(normalize=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['account_age_days','total_transactions_user','avg_amount_user','amount','shipping_distance_km']

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
cat_cols = ['country','bin_country','channel','merchant_category','promo_used','avs_match','cvv_result','three_ds_flag']

for col in cat_cols:
    print(df[col].value_counts())
    sns.countplot(data=df, x=col)
    plt.show()
```
